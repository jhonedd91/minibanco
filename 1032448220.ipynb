```python
#minibanco
saldo = 0

# Primera función - No recibe datos, no retorna datos
def imprimir_bienvenida():
    print("Bienvenido al banco XYZ")

# Segunda función - Recibe datos, no retorna datos
def realizar_consignacion():
    global saldo
    identificacion = int(input("identificación del cliente: "))
    nombre = input("nombre del cliente: ").title()
    tipo_cuenta = input("tipo de cuenta: ").upper()
    consignacion = float(input("valor a consignar: "))

    print("Su identificación es: ", identificacion)
    print("Su nombre es: ", nombre)
    print(f"Su tipo de cuenta es: {tipo_cuenta}")
    print("El valor a consignar es: ${:,.2f}".format(consignacion))

    # Línea con operadores racionales
    if consignacion > 10000:
        cobro = 4 * 1000
        print(f"El valor consignado es mayor que 10000. Se le aplicará un cobro adicional de ${cobro}.")
        saldo += consignacion - cobro
    else:
        saldo +=  consignacion

    print(f"Se ha realizado una consignación de ${consignacion:,.2f}. Su saldo actual es: ${saldo:,.2f}")

    # Línea con operadores lógicos
    if tipo_cuenta == "AHORROS" or tipo_cuenta == "AHORRO":
        print("Su cuenta es de ahorros.")
    else:
        print("Su cuenta no es de ahorros.")

# Tercera función - No recibe datos, retorna datos
def consultar_saldo():
    return saldo

# Cuarta función - Recibe datos, retorna datos
def calcular_interes(monto):
    interes = monto * 0.05
    return interes

while True:
    imprimir_bienvenida()

    # Menú con for
    opciones = ["Consultar saldo", "Realizar consignación", "Salir"]
    for i, opcion in enumerate(opciones):
        print(f"{i+1}. {opcion}")

    opcion = int(input("Ingrese una opción: "))

    if opcion == 1:
        saldo_actual = consultar_saldo()
        print(f"Su saldo actual es: ${saldo_actual:,.2f}")
    elif opcion == 2:
        realizar_consignacion()
    elif opcion == 3:
        print("Gracias por utilizar los servicios de nuestro banco.")
        break
    else:
        print("Opción no válida. Inténtelo de nuevo.")
        continue

    # Ciclo while
    while True:
        respuesta = input("¿Desea calcular el interés sobre su saldo? (s/n): ")
        if respuesta == "s":
            saldo_actual = consultar_saldo()
            interes = calcular_interes(saldo_actual)
            print(f"El interés sobre su saldo es: ${interes:,.2f}")
            break
        elif respuesta == "n":
            break
        else:
            print("Respuesta no válida. Inténtelo de nuevo.")
            continue
  
```
